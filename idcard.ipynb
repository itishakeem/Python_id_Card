```python
# prompt: increase the resolution to hd of the above card

from PIL import Image
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont


def generate_id_card(name, roll_no, distance_learning, city, center, campus, days_time, batch, image_path):
  """Generates an ID card with the provided details in HD resolution."""

  try:
    img = Image.open(image_path)
    # Resize the image for HD resolution
    img = img.resize((300, 300))  # Increased image size for HD resolution

    # Create a white background image with increased dimensions for HD resolution
    background_width = 1200  # Increased width for HD resolution
    background_height = 700  # Increased height for HD resolution
    background = Image.new('RGB', (background_width, background_height), color='white')

    # Create a green border
    border_width = 10  # Increased border width for HD resolution
    border_color = (0, 128, 0)  # Green color
    draw = ImageDraw.Draw(background)
    draw.rectangle([(0, 0), (background_width, background_height)], outline=border_color, width=border_width)

    # Paste the image onto the background (upper right corner)
    background.paste(img, (background_width - img.width - 40, 40))  # Adjust position as needed

    # Create a drawing object
    draw = ImageDraw.Draw(background)

    # Add heading "ID CARD" in green color
    font = ImageFont.load_default()
    font_size = 72  # Increased font size for HD resolution
    # Use a font available on the system
    heading_font = ImageFont.truetype("/usr/share/fonts/truetype/liberation/LiberationSans-Bold.ttf", font_size)
    draw.text((background_width // 2 - 160, 40), "ID CARD", fill=(0, 128, 0), font=heading_font)

    # Add text to the ID card with increased font size and green subheadings
    font_size = 36  # Increased font size for HD resolution
    font = ImageFont.truetype("/usr/share/fonts/truetype/liberation/LiberationSans-Regular.ttf", font_size)
    draw.text((60, 140), "Name:", fill=(0, 128, 0), font=font)  # Green subheading
    draw.text((200, 140), name, fill="black", font=font)
    draw.text((60, 200), "Roll No:", fill=(0, 128, 0), font=font)  # Green subheading
    draw.text((240, 200), roll_no, fill="black", font=font)
    draw.text((60, 260), "Distance Learning:", fill=(0, 128, 0), font=font)  # Green subheading
    draw.text((380, 260), distance_learning, fill="black", font=font)
    draw.text((60, 320), "City:", fill=(0, 128, 0), font=font)  # Green subheading
    draw.text((180, 320), city, fill="black", font=font)
    draw.text((60, 380), "Center:", fill=(0, 128, 0), font=font)  # Green subheading
    draw.text((220, 380), center, fill="black", font=font)
    draw.text((60, 440), "Campus:", fill=(0, 128, 0), font=font)  # Green subheading
    draw.text((220, 440), campus, fill="black", font=font)
    draw.text((60, 500), "Days/Time:", fill=(0, 128, 0), font=font)  # Green subheading
    draw.text((280, 500), days_time, fill="black", font=font)
    draw.text((60, 560), "Batch:", fill=(0, 128, 0), font=font)  # Green subheading
    draw.text((200, 560), batch, fill="black", font=font)

    # Add red rectangle with "Q1"
    red_rectangle_x = 40
    red_rectangle_y = 620
    red_rectangle_width = 200  # Increased width
    red_rectangle_height = 50  # Increased height
    draw.rectangle([(red_rectangle_x, red_rectangle_y), (red_rectangle_x + red_rectangle_width, red_rectangle_y + red_rectangle_height)], fill=(255, 0, 0))
    draw.text((red_rectangle_x + 80, red_rectangle_y + 15), "Q1", fill="white", font=font)

    # Add green rectangle with "WMD"
    green_rectangle_x = red_rectangle_x + red_rectangle_width + 20  # Position to the right of the red rectangle
    green_rectangle_y = red_rectangle_y  # Same Y position as red rectangle
    green_rectangle_width = 200  # Increased width
    green_rectangle_height = 50  # Increased height
    draw.rectangle([(green_rectangle_x, green_rectangle_y), (green_rectangle_x + green_rectangle_width, green_rectangle_y + green_rectangle_height)], fill=(0, 100, 0))  # Dark green color
    draw.text((green_rectangle_x + 80, green_rectangle_y + 15), "WMD", fill="white", font=font)

    # Add a line above Authorized Signature
    line_y = background_height - 140
    draw.line([(background_width - 400, line_y), (background_width - 40, line_y)], fill="black", width=4)

    # Add Authorized Signature line in the right lower corner with dark green color
    draw.text((background_width - 400, background_height - 100), "Authorized Signature", fill=(0, 100, 0), font=font)  # Dark green color

    plt.figure(figsize=(12, 7))  # Set figure size for HD resolution
    plt.imshow(background)
    plt.axis('off')
    plt.show()

  except FileNotFoundError:
    print(f"Image not found at: {image_path}")
    return


# Example usage:
generate_id_card(
    name="Anonymous",  # Changed name to Anonymous
    roll_no="PIAIC230184",
    distance_learning="No",
    city="Karachi",
    center="Bahria Auditorium",
    campus="Karsaz",
    days_time="Sunday - 09:00 AM to 01:00",
    batch="61",
    image_path="/content/unknown.jpg"  # Changed image path
)
```
